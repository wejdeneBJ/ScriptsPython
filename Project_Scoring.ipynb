```python
data1=data=data.drop('VIABILITE', axis=1)
data1=data=data.drop('NOTE_CONTRE_PARTIE', axis=1)
data1=data=data.drop('NOTE_PROJET', axis=1)
data1=data=data.drop('CLASSIFICATION_CONTRE_PARTIE', axis=1)
data1=data=data.drop('CLASSIFICATION_PROJET', axis=1)
data1=data=data.drop('NOTE_TRANSACTION', axis=1)
```

# <font color=red>Imputation des données </font>

```python



note_reac = data1['REACTIVITE_COMPORTEMENT'].fillna(0)
note_reg = data1['RISQUE_REGLEMENTAIRE'].fillna(0)
note_strat = data1['RISQUE_STRAT_COMMERCIALE'].fillna(0)
note_via = data1['NOTE_VIABILITE'].fillna(data1['NOTE_VIABILITE'].mean())
qualite_etude=data1['QUALITE_ETUDE'].fillna(0)
prise_risque_prom=data1['PRISE_RISQUE_PROMOTEUR'].fillna(0)
risque_infra=data1['RISQUE_INFRAST_LOGISTIQUE'].fillna(0)

aptitudes_gest_adm_fin=data1['APTITUDES_GESTION_ADM_FIN'].fillna(0)
aptitudes_exp_sect_act=data1['APTITUDES_EXP_SECT_ACTIVITE'].fillna(0)
aptitudes_exp_aspects_cciaux=data1['APTITUDES_EXP_ASPECTS_CCIAUX'].fillna(0)
risque_tech_couv=data1['RISQUE_TECH_COUVERTURE'].fillna(0)
risque_marche=data1['RISQUE_MARCHE'].fillna(0)
risque_appro=data1['RISQUE_APPRO'].fillna(0)
risque_concurrence=data1['RISQUE_CONCURRENCE'].fillna(0)
risque_prod=data1['RISQUE_PRODUIT'].fillna(0)

risque_assise_fi_soutien=data1['RISQUE_ASSISE_FI_SOUTIEN'].fillna(0)
risque_concentration_solva_client=data1['CONCENTRATION_SOLVA_CLIENT'].fillna(0)
risque_remboursement=data1['APTITUDE_REMBOURSEMENT'].fillna(0)
risque_mobilisation_credit=data1['MOBILISATION_CREDIT'].fillna(0)


         
```

```python
data1 = data1[['CODE_DEMANDE','RISQUE_ENG_PROMOTEUR',
             'RISQUE_MANAGERIAL',
              'RISQUE_PRODUCTION',
              'RISQUE_MARCHE_PRODUIT','RISQUE_FINANCIER'
             ]].join(note_via).join(note_reac).join(note_reg).join(note_strat).join(qualite_etude).join(prise_risque_prom).join(risque_infra).join(aptitudes_gest_adm_fin).join(aptitudes_exp_sect_act).join(aptitudes_exp_aspects_cciaux).join(risque_tech_couv).join(risque_marche).join(risque_appro).join(risque_concurrence).join(risque_prod).join(risque_assise_fi_soutien).join(risque_concentration_solva_client).join(risque_remboursement).join(risque_mobilisation_credit)
```

# <font  color='blue'> Diviser la base de données en deux tableaux $X$ (tableau d'observations) et $y$ (la variable cible) "NOTE_PROJET"</font>

```python
X= data1.drop('NOTE_VIABILITE', axis=1)

y = data1['NOTE_VIABILITE']
```

```python
X
```

# <font color=Red><font color=Red>Entrainement et evaluation d'un modèle de regression</font>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

# <font color=red>Normalisation et standarisation des variables </font>

```python
from sklearn.preprocessing import StandardScaler
sc  = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# <font color=Red>Algorithme : LinearRegression</font>


```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split 
```

```python
logreg=LinearRegression()
logreg.fit(X_train, y_train)
```

```python
print('le socre est ',logreg.score(X_train,y_train))
print('le score est',logreg.score(X_test, y_test))
```

```python
predicted = logreg.predict(X_test)
expected = y_test
print(predicted)
```
